```python
from matplotlib import pyplot as plt
import matplotlib as mpl
import matplotlib.patches
mpl.rcParams['figure.dpi'] = 300

levels, categories = pd.factorize(df['key'])
colors = [plt.cm.tab10(i) for i in levels] # using the "tab10" colormap
handles = [matplotlib.patches.Patch(color=plt.cm.tab10(i), label=c) for i, c in enumerate(categories)]

plt.scatter(df['yearfloat'], df['size'], c=colors)
plt.plot(df['yearfloat'], df['cumulative average'], color='grey', linestyle='dashed', linewidth=1, label='Cumulative mean')
plt.yscale('log')
plt.gca().set(xlabel='Publication year', ylabel='Searchspace size')
plt.xticks(np.arange(df['year'].min(), df['year'].max()+1, 1))
plt.legend(handles=handles, title='Paper')
plt.legend()
```
